```python
#order
class Oder:
    menu=['','Americano','Latte','Yogurt','Ade']
    price=[0,1500, 2500, 3000 ,2000]
    option=['','Hot','Ice']
    size=['','T','G']
    option_price=[0,0,500]
    total_price=0
    total_menu=[]
    select_menu=[]
    
    def menu_print(self, c1, c2):
        print('-'*20)
        for i in range(1,len(c1)):
            print(i, '.', c1[i], ':', c2[i],'원')
        print('-'*20)
            
    def menu_select(self):
        self.menu_print(self.menu, self.price)
        choice = int(input('메뉴 번호를 선택해주세요  :  '))
        
        while choice != 0:

            cal_price = self.price[choice]
            
            if choice == 1 or choice == 2:
                self.menu_print(self.option, self.option_price)
                o = int(input('옵션를 선택해주세요. 추가 금액 있을 수 있습니다.  :  '))
                cal_price +=self.option_price[o] 
            else:
                o = 2
                print('차가운 음료만 제공합니다.  :  ')

            self.menu_print(self.size, self.option_price)
            s = int(input('사이즈를 선택해주세요  :  '))
            cal_price += self.option_price[s]

            n = int(input('몇 잔을 원하시나요?  :  '))
            cal_price = cal_price*n
            self.total_price += cal_price
            select_menu = [self.menu[choice], self.option[o], self.size[s]+'사이즈', str(n)+'잔']
            self.total_menu.append(select_menu)
            
            print('-'*20)
            choice =  int(input('메뉴추가? 원하면 메뉴 번호 선택, 아니면 0  :  '))
            
           
        print('-'*20)
        print('주문하신 음료는 다음과 같습니다.')
        print(self.total_menu)
        print('결제하실 금액은 ', self.total_price, '원입니다.')

        
customer=Oder()
customer.menu_select()

```
