#### 1. 아래 내용을 참고하여 Player 클래스를 만드세요.
1. Player 클래스 생성
2. player는 닉네임, HP, MP를 가짐(속성)
3. player는 다른 player을 상대로 공격, 스킬, 방어를 가능(함수로 구현)
4. (선택사항)속성이나 player 행동에 대한 함수를 자유롭게 추가해도 됨

````python
# 정해진 답은 없음, 클래스를 이해하고자라는 의도로 만든 문제
# 아래 코드는 위 참고 내용의 1~3번만 구현한 코드임

# Solution
class Player:
    def __init__(self, name, hp, mp):
        self.name = name
        self.hp = hp
        self.mp = mp


    def attack(self, target):
        target.hp = target.hp - 10
        if (target.hp <= 0):
            print("%s player가 공격하여 %s player를 죽였습니다!" % (self.name, target.name))
        else:
            print("%s player가 공격하여 %s player의 HP가 %d이 되었습니다" % (self.name, target.name, target.hp))


    def skill(self, target):
        target.hp = target.hp - 100
        if (target.hp <= 0):
            print("%s player가 skill을 사용하여 %s player를 죽였습니다!" % (self.name, target.name))
        else:
            print("%s player가 공격하여 %s player의 HP가 %d이 되었습니다" % (self.name, target.name, target.hp))

    # 방어 대신 힐로 구현
    def heal(self, target):
        target.hp = target.hp + 50
        print("%s player가 힐을 사용하여 %s player의 HP가 %d이 되었습니다" % (self.name, target.name, target.hp))


Archer = Player('활내골라스', 100, 50)
Warrior = Player('타락파워전사', 100, 50)
Priest = Player('걸어다니는포션', 100, 50)

Archer.skill(Warrior)
Priest.attack(Archer)
````

````python
# Solution
import random

class Account:
    # 클래스 변수
    account_count = 0

    def __init__(self, name, balance):
        self.deposit_count = 0  # 입금횟수 초기화
        self.deposit_log = []   # 입금내역
        self.withdraw_log = []  # 출금내역

        # 빠뜨린 속성 추가 구현
        # ...
        self.bank = "SC은행"
        self.name = name
        self.balance = balance

        # 랜덤하게 개좌번호 생성을 구현
        # ...
        num1 = str(random.randint(100,999))
        num2 = ""
        for _ in range(2):
            num2 = num2 + str(random.randint(0,9))
        num3 = ""
        for _ in range(6):
            num3 = num3 + str(random.randint(0,9))

        self.account_number = num1 + '-' + num2 + '-' + num3
        Account.account_count += 1


    @classmethod
    def get_account_num(cls):
        print(cls.account_count)


    def deposit(self, amount):
        if amount >= 1:
            self.deposit_log.append(amount)
            self.balance += amount

            self.deposit_count += 1
        if self.deposit_count % 5 == 0:   # 5, 10, 15 ...
            self.balance *= 1.01            # 이자


    def withdraw(self, amount):
        if self.balance >= amount:
            self.withdraw_log.append(amount)
            self.balance -= amount


    def display_info(self):
        # 정보출력을 구현
        # ...
        print(f"은행이름: {self.bank}")
        print(f"예금주: {self.name}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔고: {self.balance:,d}")


    def deposit_history(self):
        # 예금 히스토리 출력을 구현
        for i in self.deposit_log:
            print(i)


    def withdraw_history(self):
        # 출금 히스토리 출력을 구현
        for j in self.withdraw_log:
            print(j)


kim = Account("김민수", 100)
print("은행이름:", kim.bank)
print("계좌번호:", kim.account_number)
print('-'*20)

kim = Account("김민수", 100)
print("총 계좌수:", Account.account_count)
lee = Account("이민수", 100)
print("총 계좌수:", Account.account_count)
print('-'*20)

kim = Account("김민수", 100)
lee = Account("이민수", 100)
Account.get_account_num()
print('-'*20)

kim = Account("김민수", 100)
kim.deposit(1)
print(kim.balance)
kim.deposit(100)
print(kim.balance)
print('-'*20)

p = Account("파이썬", 10000)
p.display_info()
print('-'*20)

kim = Account("김민수", 100)
kim.deposit(100)
kim.withdraw(90)
print(kim.balance)
print('-'*20)

p = Account("파이썬", 10000)
p.deposit(100)
p.deposit(200)
p.deposit(300)
p.deposit_history()
print('-'*20)

p.withdraw(100)
p.withdraw(200)
p.withdraw_history()
print('-'*20)
````
